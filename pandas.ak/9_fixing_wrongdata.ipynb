# Wrong Data:
"Wrong data" does not have to be "empty cells" or "wrong format", it can just be wrong, like if someone registered "199" instead of "1.99".

```python
# loading and reading the original dataframe.
import pandas as pd
ak = pd.read_csv("dirtydata.csv")
print(ak.to_string())
```

```python
# here we will set "Duration" = 45 in row 7.
import pandas as pd
ak=pd.read_csv("dirtydata.csv")
ak.loc[7,"Duration"]=45
print(ak.to_string())
```

# for larger data set:
now here we will loop throgh all the values in "Duration" column,if the values is higher than 120 than set it to 120.

```python
import pandas as pd
ak = pd.read_csv("dirtydata.csv")
for i in ak.index:
    if ak.loc[i,"Duration"]>120:
        ak.loc[i,"Duration"]=120
print(ak.to_string())
```

# remove the rows in wrong data in large dataset

```python
import pandas as pd
ak = pd.read_csv("dirtydata.csv")
for i in ak.index:
    if ak.loc[i,"Duration"]>120:
        ak.drop(i, inplace=True)
print(ak.to_string())
```
